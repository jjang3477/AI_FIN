```python
#### 구간별 변동 데이터 추출 프로그램 ###
#### time define
from datetime import datetime as dt
from datetime import date, timedelta
```

```python
### 시작 / 끝 날짜 지정
now = dt.today().date()
### 30일 전 , 
start = now - timedelta(days=30)
start_date = start.strftime('%Y%m%d')
end_date = now.strftime('%Y%m%d')
```

```python
### 기간별 시작 / 끝 날짜 return 함수 
def date_define(num_day):
    start = now - timedelta(days=num_day)
    start_date = start.strftime('%Y%m%d')
    end_date = now.strftime('%Y%m%d')
    return start_date,end_date

date_180 = date_define(180)
date_60 = date_define(60)
date_30 = date_define(30)
date_7 = date_define(10)
```

```python
date_60[0]
df_6month = stock.get_market_price_change_by_ticker(date_180[0], date_180[1])
df_3month = stock.get_market_price_change_by_ticker(date_60[0], date_60[1])
df_1month = stock.get_market_price_change_by_ticker(date_30[0], date_30[1])
df_1week = stock.get_market_price_change_by_ticker(date_7[0], date_7[1])

#df_1month = stock.get_market_price_change_by_ticker("20210401", "20210530")
#df_1weeks = stock.get_market_price_change_by_ticker("20210515", "20210530")
```

```python
#df_6month
df_6month = df_6month.sort_values(by=['등락률'], axis=0,ascending=False)
df_3month = df_3month.sort_values(by=['등락률'], axis=0,ascending=False)
df_1month = df_1month.sort_values(by=['등락률'], axis=0,ascending=False)
df_1week = df_1week.sort_values(by=['등락률'], axis=0,ascending=False)

df_6month
```

```python
### merge 
## https://yganalyst.github.io/data_handling/Pd_12/
    
merge_outer1 = pd.merge(df_6month,df_3month, how='left',on='티커')
merge_outer2 = pd.merge(df_1month,df_1week, how='left',on='티커')
merge_outer = pd.merge(merge_outer1,merge_outer2, how='left',on='티커')

#df_3month.set_index('티커', inplace=True)
```

```python
#merge_outer.head(1)
merge_outer.columns
```

```python
## 특정 컬럼 추출 및 이름 변경 ## 
df_info= merge_outer[['종목명_x_x','등락률_x_x','등락률_y_x','등락률_x_y','등락률_y_y']]
df_info=df_info.rename(columns={'종목명_x_x':'종목명','등락률_x_x': '6달_등락률','등락률_y_x': '3달_등락률',
                                '등락률_x_y': '1달_등락률','등락률_y_y': '1주일_등락률'})
```

```python
### 컬럼 연산 
df_info.head(1)
df_info['6달_비율'] = df_info['6달_등락률']/100
df_info['6달_1일_비율'] = df_info['6달_비율']/180
df_info['3달_비율'] = df_info['3달_등락률']/100
df_info['3달_1일_비율'] = df_info['3달_비율']/180

df_info['1달_비율'] = df_info['1달_등락률']/100
df_info['1달_1일_비율'] = df_info['1달_비율']/180
df_info['1주일_비율'] = df_info['1주일_등락률']/100
df_info['1주일_1일_비율'] = df_info['1주일_비율']/180

df_info= df_info[['종목명','6달_비율','6달_1일_비율','3달_비율','3달_1일_비율',
                  '1달_비율','1달_1일_비율','1주일_비율','1주일_1일_비율']]
```

```python
### 숫자 포멧 변경
pd.set_option('display.float_format', '{:.2f}'.format) # 항상 float 형식으로
#pd.set_option('display.float_format', '{:.2e}'.format) # 항상 사이언티픽
#pd.set_option('display.float_format', '${:.2g}'.format)  # 적당히 알아서
pd.options.display.float_format = '{:.2f}'.format
df_info.head(30)
```
